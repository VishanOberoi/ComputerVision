```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image_path = '/content/fruits.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
cv2_imshow(image)
```

```python

edges = cv2.Canny(image, threshold1=98, threshold2=100)

cv2_imshow(edges)
```

```python
#Finding contours

contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(contours)

```

```python
contours[0] #these are row and col coordinates
```

```python
image = cv2.imread('/content/fruits.png')
image_contours = image.copy()
cont_wthout_morph = cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)
plt.imshow(cont_wthout_morph)
```

# Same with Morphological Ops

```python

se = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
#Now dilate then erode
dilate_image = cv2.dilate(edges, se, iterations = 1)
erode_image = cv2.erode(dilate_image, se, iterations = 1)
cv2_imshow(erode_image)
```

```python
contours_with_morph, _ = cv2.findContours(erode_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(contours_with_morph)
```

```python
image = cv2.imread('/content/fruits.png')
image_contours = image.copy()
morph_drawn = cv2.drawContours(image_contours, contours_with_morph, -1, (0, 255, 0), 2)
plt.imshow(morph_drawn)
```

```python
plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(cont_wthout_morph)
plt.title("Before Morphology")

plt.subplot(1,3,3)
plt.imshow(morph_drawn)
plt.title("After Morphology")
```

```python
#For the phone Image
image_path = '/content/cellphone.JPG'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(image, threshold1=75, threshold2=200)
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
len(contours)
image = cv2.imread(image_path)
image_contours = image.copy()
cont_wthout_morph = cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)
se = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
dilate_image = cv2.dilate(edges, se, iterations = 1)
erode_image = cv2.erode(dilate_image, se, iterations = 1)
contours_with_morph, _ = cv2.findContours(erode_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
morph_drawn = cv2.drawContours(image_contours, contours_with_morph, -1, (0, 255, 0), 2)
plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(cont_wthout_morph)
plt.title("Before Morphology")

plt.subplot(1,3,3)
plt.imshow(morph_drawn)
plt.title("After Morphology")
```

#Conclusion


* If we reduce the lower threshold keeping the upper constant then more edges are detected by the Canny Edge Detection Algorithm
* If we increase upper threshold value keeping the lower same means less edges will be detected
* If all points are considered instead of approximating then the contours will be more accurcate
* Morphological ops :
** Dilation used to fill the gap between edge points detected by canny edge detection algo. If number of iterations increased then thickness of edges increase.
** If dilated edges are then followed by a number of iterations of erosion, then the thickness of the edges can be controlled to suit the requirement of the user.
** Morph ops are followed by contour detection which gives us a better quality of edges.
