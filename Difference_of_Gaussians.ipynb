```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
from skimage.io import imread, imshow
from skimage import data
from skimage.color import rgb2gray

image = data.astronaut()
imshow(image)
```

```python
#Filter 1
filter_1_size = (5, 5)
filter_1_sigma = 2.0
filtered_image_1 = cv2.GaussianBlur(image, filter_1_size, filter_1_sigma)
```

```python
#Filter 2
filter_2_size = (5, 5)
filter_2_sigma = 5
filtered_image_2 = cv2.GaussianBlur(image, filter_2_size, filter_2_sigma)
```

```python
dog_image = filtered_image_2 - filtered_image_1
```

```python
plt.imshow(dog_image)
```

# Resizing

```python
from skimage.transform import resize
image = data.astronaut()
image_resized = resize(image, (image.shape[0] // 2, image.shape[1] // 2),
                       anti_aliasing=True)
plt.imshow(image_resized)
```

```python
#Filter 1
filter_1_size = (5, 5)
filter_1_sigma = 0.1
filtered_image_1 = cv2.GaussianBlur(image_resized, filter_1_size, filter_1_sigma)
#Filter 2
filter_2_size = (5, 5)
filter_2_sigma = 0.
filtered_image_2 = cv2.GaussianBlur(image_resized, filter_2_size, filter_2_sigma)

dog_image = filtered_image_1 - filtered_image_2
plt.imshow(dog_image)
```
