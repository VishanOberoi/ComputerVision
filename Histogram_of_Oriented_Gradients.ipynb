#Taking any image


```python


from skimage.io import imread, imshow
from skimage import data
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

image = data.coffee()
imshow(image)
```

# Cropping the image


```python

cropped_image = image[150:250, 400:450]
imshow(cropped_image)
```

# Resize the image to 128x64


```python

image = cv2.resize(cropped_image, (64, 128))
imshow(image)
```

#Compute the HOG

```python
from skimage.feature import hog
from skimage import data, exposure

fd_1, hog_image_1 = hog(image, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel= True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Input image')

# Rescale histogram for better display
hog_image_rescaled_1 = exposure.rescale_intensity(hog_image_1, in_range=(0, 10))

ax2.axis('off')
ax2.imshow(hog_image_rescaled_1, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
plt.show()
```

# Different part of image

```python


from skimage.io import imread, imshow
from skimage import data
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

image = data.coffee()
cropped_image = image[150:250, 80:130]
image = cv2.resize(cropped_image, (64, 128))
from skimage.feature import hog
from skimage import data, exposure

fd_3, hog_image_3 = hog(image, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True, multichannel= True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Input image')

# Rescale histogram for better display
hog_image_rescaled_3 = exposure.rescale_intensity(hog_image_3, in_range=(0, 10))

ax2.axis('off')
ax2.imshow(hog_image_rescaled_3, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
plt.show()
```

#Conclusion


* From the given image, two objects are extracted as patches
* The same patch is resized to 128x64 columns, or mulitples of the same.
* The size of each cell is 8x8 px, and each block contains 2x2 number of cells
* Number of bins = 9

* For the above set of parameters, visualisation of hog shows the brighter parts on edges of the object, and dark pixels where the image has pixels with constant intensity

* Length of feature vector is 3780. (7*15*2*2*9)


* If cell size is increased to 16x16, HOG of image shows broader details, as multiple local features combined into one cell and fine detail are combined into broader details.
* If cell pixel size 16x16 changed to 8x8 , length of the feature vector changes from 3780 to 756. Since, less features are there in the resized image.
* Euclidean distance between feature vector of 2 images which are similar proves to be 6.9.
* Distance is large if objects are dissimlar, and less if objects are similar.


Therfore, we can conclude that visualisation of HOG can be used to detect the edges and feature vectors can be used to compare two objects.
