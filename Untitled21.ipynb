```python
v = cv2.VideoCapture('/content/traffic.avi')

f = []
k = 0
while True:
  r,frame = v.read()
  if frame is not None:
    f.append(frame)
    k = k + 1
  else:
    break

frames = np.array(f)

```

```python
shape = frames.shape
shape
```

```python
#Take each frame and pass it to the model

gmm = GaussianMixture(n_components=2)
shape = frames.shape
background = np.zeros(shape[1:])



for rows in range(shape[1]):
  for cols in range(shape[2]):
    for chanels in range(shape[3]):
      pix = frames[:, rows,cols,chanels]
      pix  = pix.reshape(pix.shape[0],1)
      gmm.fit(pix)
      means = gmm.means_
      weights = gmm.weights_
      index = np.argmax(weights)
      background[rows][cols][chanels] = int(means[index])
```

```python
cv2.imwrite("After.png", background)
```

```python
image = frames[10,:,:,:]

foreground = image - background

f_gray = rgb2gray(foreground)

resized_image = cv2.resize(f_gray, (500, 500))

cv2_imshow(resized_image)
```

```python
v = cv2.VideoCapture('/content/traffic.avi')

f = []
k = 0
while True:
  r,frame = v.read()
  if frame is not None:
    f.append(frame)
    k = k + 1
  else:
    break

frames = np.array(f)




#Take each frame and pass it to the model

gmm = GaussianMixture(n_components=3)
shape = frames.shape
background = np.zeros(shape[1:])



for rows in range(shape[1]):
  for cols in range(shape[2]):
    for chanels in range(shape[3]):
      pix = frames[:, rows,cols,chanels]
      pix  = pix.reshape(pix.shape[0],1)
      gmm.fit(pix)
      means = gmm.means_
      weights = gmm.weights_
      index = np.argmax(weights)
      background[rows][cols][chanels] = int(means[index])
image = frames[10,:,:,:]

foreground = image - background

f_gray = rgb2gray(foreground)

resized_image = cv2.resize(f_gray, (500, 500))

cv2_imshow(resized_image)
```
